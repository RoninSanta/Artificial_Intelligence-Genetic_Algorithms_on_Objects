```python
import random
import time
import matplotlib.pyplot as plt

# Genetic Algorithm Parameters
mutation_rate = 0.01  # Probability of mutation
mutation_range = 0.1  # Range of mutation
population_size = 100
num_generations = 100

# Initialize population
population = []
for _ in range(population_size):
    chromosome = [random.uniform(0, 1) for _ in range(10)]  # Example chromosome length of 10
    population.append(chromosome)

# Evaluation function (you need to define your own fitness function)
def evaluate_fitness(chromosome):
    # Calculate fitness based on the specific problem/task
    return sum(chromosome)

# Track evolution progress
mean_fitness_values = []
max_fitness_values = []

# Variables for tracking metrics
num_individuals_evaluated = 0
start_time = time.time()

# Main evolution loop
for generation in range(num_generations):
    # Evaluate fitness for each chromosome in the population
    fitness_values = [evaluate_fitness(chromosome) for chromosome in population]

    # Record mean and maximum fitness for the current generation
    mean_fitness = sum(fitness_values) / len(fitness_values)
    mean_fitness_values.append(mean_fitness)
    max_fitness = max(fitness_values)
    max_fitness_values.append(max_fitness)

    # Track the number of individuals evaluated
    num_individuals_evaluated += population_size

    # Selection
    selected_population = random.choices(population, weights=fitness_values, k=population_size)

    # Reproduction (crossover and mutation)
    new_population = []
    for _ in range(population_size):
        parent1, parent2 = random.choices(selected_population, k=2)
        offspring = []
        for gene_index in range(len(parent1)):
            if random.random() < mutation_rate:
                mutated_value = parent1[gene_index] + random.uniform(-mutation_range, mutation_range)
                offspring.append(mutated_value)
            else:
                offspring.append(parent1[gene_index])
        new_population.append(offspring)

    # Update population with the new generation
    population = new_population

    # Check convergence condition (e.g., based on mean fitness improvement)
    if len(mean_fitness_values) > 1 and abs(mean_fitness_values[-1] - mean_fitness_values[-2]) < 0.001:
        convergence_time = time.time() - start_time
        break

# Print metrics
print("Number of individuals evaluated:", num_individuals_evaluated)
print("Number of generations:", generation + 1)
print("Convergence time:", convergence_time)
print("Maximum Fitness:", max(max_fitness_values))

# Plot the evolution progress
plt.plot(range(len(mean_fitness_values)), mean_fitness_values)
plt.xlabel('Generation')
plt.ylabel('Mean Fitness')
plt.title('Evolution Progress')
plt.show()
```
